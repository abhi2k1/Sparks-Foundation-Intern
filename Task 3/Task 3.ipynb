4. Relationship Analysis

```python
sns.pairplot(data)
```

```python
plt.pie(data['Ship Mode'].value_counts(),labels=data['Ship Mode'].unique(),autopct ='% 1.1f %%',shadow=True,startangle=90)
plt.title("Shipment Mode")
plt.show()
```

```python
plt.pie(data['Segment'].value_counts(),labels=data['Segment'].unique(),autopct ='% 1.1f %%',shadow=True,startangle=90)
plt.title("Segment")
plt.show()
```

```python
plt.pie(data['Region'].value_counts(),labels=data['Region'].unique(),autopct ='% 1.1f %%',shadow=True,startangle=90)
plt.title("Region")
plt.show()
```

```python
plt.pie(data['Category'].value_counts(),labels=data['Category'].unique(),autopct ='% 1.1f %%',shadow=True,startangle=90)
plt.title('Category')
plt.show()
```

```python
plt.subplots(figsize=(16,8))
sns.countplot(x=data['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.subplots(figsize=(16,8))
sns.countplot(x=data['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```
